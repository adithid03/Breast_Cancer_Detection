# Outlier Detection: Using IQR

Retrieves the 25th percentile (Q1) for each feature. Retrieves the 75th percentile (Q3) for each feature. IQR is the difference between the third quartile (Q3) and first quartile (Q1).


```python
# Calculate the IQR for each feature
Q1 = pd.DataFrame(X_scaled, columns=X.columns).quantile(0.25)  # First quartile (25th percentile)
Q3 = pd.DataFrame(X_scaled, columns=X.columns).quantile(0.75)  # Third quartile (75th percentile)
IQR = Q3 - Q1  # IQR is the difference between the 75th and 25th percentiles
```

```python
# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR  # Anything below this is considered an outlier
upper_bound = Q3 + 1.5 * IQR  # Anything above this is considered an outlier
```

```python
# Identify outliers by checking if any values are outside the lower and upper bounds
outliers = ((pd.DataFrame(X_scaled, columns=X.columns) < lower_bound) | (pd.DataFrame(X_scaled, columns=X.columns) > upper_bound))
```

```python
# Get the number of outliers for each column
num_outliers_per_column = outliers.sum()
```

```python
# Display the number of outliers for each feature column
print("Number of outliers per feature column:\n", num_outliers_per_column)
```

# Implementing Adaboost Algorithm

## Import necessary libraries

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
```

## Define hyperparameters for AdaBoost

```python
n_estimators = 50           # Number of weak learners (decision stumps)
learning_rate = 1.0         # Learning rate for AdaBoost
```

## Initialize the AdaBoost classifier with a Decision Tree as the base estimator

```python
base_estimator = DecisionTreeClassifier(max_depth=1)  # Using a weak learner
adaboost_clf = AdaBoostClassifier(
    estimator=base_estimator, 
    n_estimators=n_estimators, 
    learning_rate=learning_rate, 
    random_state=42
)
```

## Fit the AdaBoost classifier on the training data and Make predictions on the test set

```python
# Fit the AdaBoost classifier on the training data
adaboost_clf.fit(X_train_smote, y_train_smote)
```

```python
# Make predictions on the test set
y_pred = adaboost_clf.predict(X_test)
```

# Evaluating Model Performance

## Accuracy

```python
# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy with outliers: {accuracy:.2f}")
```

## Precision

```python
# Calculate precision
precision = precision_score(y_test, y_pred, pos_label='M')  # Assuming 'M' for Malignant
print(f"Precision with outliers: {precision:.2f}")
```

## Recall

```python
# Calculate recall
recall = recall_score(y_test, y_pred, pos_label='M')
print(f"Recall with outliers: {recall:.2f}")
```

## F1-score

```python
# Calculate F1-score
f1 = f1_score(y_test, y_pred, pos_label='M')
print(f"F1 Score with outliers: {f1:.2f}")
```

##  Confusion Matrix

```python
conf_matrix = confusion_matrix(y_test, y_pred, labels=['M', 'B'])  # Assuming 'M' for Malignant, 'B' for Benign
print("Confusion Matrix with outliers:")
print(conf_matrix)
```
